```python
import pandas as pd
```

```python
df35 = pd.read_csv('scores_gpt35.csv')
df35
```

```python
df4 = pd.read_csv('scores_gpt4.csv')
df4
```

```python
df35['average_score'] = df35[['interests_score', 'budget_score', 'Travel_Score', 'redundency_score']].mean(axis=1)
df35 = df35[['location', 'interests_score', 'budget_score', 'Travel_Score', 'redundency_score', 'average_score']]
df35
```

```python
df4['average_score'] = df4[['interests_score', 'budget_score', 'Travel_Score', 'redundency_score']].mean(axis=1)
df4 = df4[['location', 'interests_score', 'budget_score', 'Travel_Score', 'redundency_score', 'average_score']]
df4
```

```python
import matplotlib.pyplot as plt

# List of columns to compare
columns_to_compare = ['interests_score', 'budget_score', 'Travel_Score', 'redundency_score', 'average_score']

# Plotting histograms for each column
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 12))
axes = axes.flatten()

for i, column in enumerate(columns_to_compare):
    axes[i].hist(df4[column], alpha=0.5, label='df4', bins=10, color='blue')
    axes[i].hist(df35[column], alpha=0.5, label='df35', bins=10, color='green')
    axes[i].set_title(column)
    axes[i].legend(loc='upper right')

# Removing any empty subplots
if len(columns_to_compare) < len(axes):
    for j in range(len(columns_to_compare), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# List of columns to compare
columns_to_compare = ['interests_score', 'budget_score', 'Travel_Score', 'redundency_score', 'average_score']

# Calculate means
means_df4 = df4[columns_to_compare].mean()
means_df35 = df35[columns_to_compare].mean()

# Bar width
bar_width = 0.35

# Indexes for the groups
index = np.arange(len(columns_to_compare))

# Plotting the bar graph
fig, ax = plt.subplots(figsize=(12, 8))

# Bars for df4
bars1 = ax.bar(index, means_df4, bar_width, label='GPT-4', color='blue')

# Bars for df35
bars2 = ax.bar(index + bar_width, means_df35, bar_width, label='GPT-3.5', color='purple')

# Adding labels, title, and legend
ax.set_xlabel('Metrics')
ax.set_ylabel('Mean Scores')
ax.set_title('Comparison of Mean Scores Between GPT-4 and GPT-3.5')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(columns_to_compare)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
